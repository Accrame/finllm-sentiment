# Prompt Format Experiments

Trying different prompt templates to see what works best for SFT.


## Template 1: Simple classification


```python
template_v1 = '''Analyze the sentiment of this financial text.

Text: {text}

Sentiment (positive/negative/neutral):'''

# This is the basic format, works ok but model sometimes
# outputs extra text after the label
```

## Template 2: With reasoning


```python
template_v2 = '''Analyze the sentiment of this financial text. Provide your reasoning.

Text: {text}

Analysis:
Sentiment: '''

# This format gets better reasoning but harder to parse
```

## Template 3: JSON output


```python
template_v3 = '''Analyze the sentiment of this financial text.

Text: {text}

Respond with JSON: {\"sentiment\": \"positive/negative/neutral\", \"confidence\": 0.0-1.0}

Response:'''

# JSON format is cleanest for downstream use
# Going with v1 for training, v3 for inference
```

## Decision
- Use template v1 for SFT training (simple, consistent)
- Use template v3 at inference time for structured output
- v2 is interesting but makes eval harder
